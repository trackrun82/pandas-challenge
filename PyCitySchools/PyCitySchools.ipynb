```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Files to Load
school_data = os.path.join("Resources", "schools_complete.csv")
student_data = os.path.join("Resources","students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)

# Combine the data into a single dataset.  
school_total_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
school_total_df.head()
```

```python
#Amend the original df to include school size
#Create bins and bin labels for school size column
school_size_bins = [0, 1000, 2000, 5000]
school_size_labels = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']
```

```python
#Bin the school size column and add to original df
school_total_df['School Size'] = pd.cut(school_total_df['size'],
                                                     school_size_bins, labels=school_size_labels)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools

#Extract school column
school_list = school_total_df['school_name']

#Form unique list of schools
unique_school_list = school_list.unique()

#Show length of unique schools array
total_schools = len(unique_school_list)
```

```python
#Calculate the total number of students

#Extract student name column
student_list = school_total_df['student_name']

#Count total students
total_students = student_list.count()
```

```python
#Calculate the total budget

#Extract budget column
budget_list = school_total_df['budget']

#Form unique list of budgets
unique_budget_list = budget_list.unique()

#Sum budgets to form total
total_budget = unique_budget_list.sum()
```

```python
#Calculate the average math score

#Extract math score column
math_list = school_total_df['math_score']

#Calculate the mean score
avg_math = math_list.mean()
```

```python
#Calculate the average reading score

#Extract reading score column
reading_list = school_total_df['reading_score']

#Calculate the mean score
avg_reading = reading_list.mean()
```

```python
#Calculate the percentage of students with a passing math grade

#Extract students who had >=70% (passing) math score
students_pass_math = school_total_df.loc[math_list >= 70]

#Count number of students who passed
total_pass_math = students_pass_math['student_name'].count()

#Calculate % passing math
percent_pass_math = total_pass_math/total_students*100
```

```python
#Calculate the percentage of students with a passing reading grade

#Extract students who had >=70% (passing) reading score
students_pass_reading = school_total_df.loc[reading_list >= 70]

#Count number of students who passed
total_pass_reading = students_pass_reading['student_name'].count()

#Calculate % passing reading
percent_pass_reading = total_pass_reading/total_students*100
```

```python
#Calculate the percentage of students who passed both reading and math

#Extract students who had >=70% (passing) reading and math score
students_pass_math_and_reading = school_total_df.loc[(math_list >= 70) & (reading_list >= 70)]

#Count number of students who passed
total_pass_both = students_pass_math_and_reading['student_name'].count()

#Calculate % passing both
percent_pass_both = total_pass_both/total_students*100
```

```python
#Create DataFrame to hold the above results
district_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                                    "Total Students": [total_students],
                                    "Total Budget": [total_budget],
                                    "Average Math Score": [avg_math],
                                    "Average Reading Score": [avg_reading],
                                    "% Passing Math": [percent_pass_math],
                                    "% Passing Reading": [percent_pass_reading],
                                    "% Overall Passing": [percent_pass_both]})
```

```python
#Give the dataframe better readability
district_summary_df['Total Students'] = district_summary_df['Total Students'].map('{:,}'.format)
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map('${:,.2f}'.format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Group set by schools
grouped_school_df = school_total_df.groupby('school_name')
```

```python
#Get total student count based on school
total_student_count = school_total_df['school_name'].value_counts()

#Return unique values of School Type and Budget
school_type = grouped_school_df['type'].describe()['top']
budget = grouped_school_df['budget'].describe()['mean']

#Calculate Budget per student
budget_perstudent = budget/total_student_count
```

```python
#Calculate average math score per school

#Extract math score column from grouped df
math_score_school = grouped_school_df['math_score']

#Calculate mean
avg_math_school = math_score_school.mean()
```

```python
#Calculate average reading score per school

#Extract reading score column from grouped df
reading_score_school = grouped_school_df['reading_score']

#Calculate mean
avg_reading_school = reading_score_school.mean()
```

```python
#Calculate % Passing for math scores >=70%

#Group math passing df by school name
grouped_math_passing = students_pass_math.groupby('school_name')

#Count total number by school
grouped_math_school = grouped_math_passing['student_name'].count()

#Calculate % passing for math
percent_pass_math_school = grouped_math_school/total_student_count*100
```

```python
#Calculate % Passing for reading scores >=70%

#Group reading passing df by school name
grouped_reading_passing = students_pass_reading.groupby('school_name')

#Count total number by school
grouped_reading_school = grouped_reading_passing['student_name'].count()

#Calculate % passing for reading
percent_pass_reading_school = grouped_reading_school/total_student_count*100
```

```python
#Calculate % Passing for both math and reading scores >=70%

#Group both passing df by school name
grouped_both_passing = students_pass_math_and_reading.groupby('school_name')

#Count total number by school
grouped_both_school = grouped_both_passing['student_name'].count()

#Calculate % passing for both
percent_pass_both_school = grouped_both_school/total_student_count*100
```

```python
#Create DataFrame to hold the above results
summary_df = pd.DataFrame({'School Type':school_type,
                           'Total Students':total_student_count,
                           'Total School Budget':budget,
                           'Per Student Budget':budget_perstudent,
                           'Avg Math Score':avg_math_school,
                           'Avg Reading Score':avg_reading_school,
                           '% Passing Math':percent_pass_math_school,
                           '% Passing Reading':percent_pass_reading_school,
                           '% Overall Passing':percent_pass_both_school})
```

```python
#Perserve original dataframe and Give the new dataframe better readability
summary_df_better_read = summary_df.copy()
summary_df_better_read['Total School Budget'] = summary_df_better_read['Total School Budget'].map('${:,.2f}'.format)
summary_df_better_read['Per Student Budget'] = summary_df_better_read['Per Student Budget'].map('${:.2f}'.format)
summary_df_better_read
```
